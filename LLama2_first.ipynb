```python
from transformers import AutoTokenizer

model = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model,use_auth_token = True)
```

```python
from transformers import pipeline
import torch
```

```python
llama_pipeline = pipeline(
    "text-generation",
    model = model,
    torch_dtype = torch.float16,
    device_map = "auto",
)
```

```python
def get_llama_response(prompt:str) -> None:
  sequence = llama_pipeline(
      prompt,
      do_sample = True,
      top_k = 10,
      num_return_sequences = 1,
      eos_token_id = tokenizer.eos_token_id,
      max_length = 256,
  )
  print("Chat:", sequence[0]['generated_text'])
```

```python
prompt = 'Tell me about the Earth.'
get_llama_response(prompt)
# to just know that does it have simple facts clear
```

```python
prompt = 'What are the neighbouring countries of India?'
get_llama_response(prompt)
# to know how much it knows in depth
```

```python
prompt = 'Till what exact date you have knowledge about the world?'
get_llama_response(prompt)
```
